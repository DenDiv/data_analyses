```python
import numpy as np
```

```python
with open('train.txt') as file:
    text = file.readlines()
```

```python
see_d = {}
buy_d = {}
for line in text:
    new_arr = line[:-1].split(';')
    see_ids = new_arr[0].split(',')
    buy_ids = new_arr[1].split(',')
    
    if buy_ids[0] == '':
        continue
        
    for see_id in see_ids:
        see_d.setdefault(see_id, 1)
        see_d[see_id] += 1
        
    for buy_id in buy_ids:
        buy_d.setdefault(buy_id, 1)
        buy_d[buy_id] += 1
```

```python
sorted_see_d = {k: v for k, v in sorted(see_d.items(), key=lambda item: item[1], reverse= True)}
sorted_see_d
```

```python
sorted_buy_d = {k: v for k, v in sorted(buy_d.items(), key=lambda item: item[1], reverse= True)}
sorted_buy_d
```
