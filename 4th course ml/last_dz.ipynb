```python
import numpy as np
import pandas as pd
from scipy import stats
import scipy
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('churn_analysis.csv')
df.head()
```

```python
contr_gr = df[df['treatment'] == 1]
```

```python
st_names = np.unique(df.state)
```

```python
cr_tab = pd.crosstab(contr_gr['state'],contr_gr['churn'])
cr_tab
```

```python
counter = 0
for f_name in range(len(st_names)):
    for l_name in range(f_name,len(st_names)):
        tabl = np.zeros((2,2))
        tabl[0,:] = cr_tab.loc[st_names[f_name]]
        tabl[1,:] = cr_tab.loc[st_names[l_name]]
        try:
            if scipy.stats.chi2_contingency(tabl, correction=False)[1] < 0.05:
                counter += 1
        except Exception:
            print('err')
counter
```

```python
counter = 0
for f_name in range(len(st_names)):
    for l_name in range(f_name,len(st_names)):
        tabl = np.zeros((2,2))
        tabl[0,:] = cr_tab.loc[st_names[f_name]]
        tabl[1,:] = cr_tab.loc[st_names[l_name]]
        try:
            if scipy.stats.chi2_contingency(tabl, correction=True)[1] < 0.05:
                counter += 1
        except Exception:
            print('err')
counter
```

```python
counter = 0
for f_name in range(len(st_names)):
    for l_name in range(f_name,len(st_names)):
        tabl = np.zeros((2,2))
        tabl[0,:] = cr_tab.loc[st_names[f_name]]
        tabl[1,:] = cr_tab.loc[st_names[l_name]]
        if scipy.stats.fisher_exact(tabl)[1] < 0.05:
            counter += 1
counter
```
